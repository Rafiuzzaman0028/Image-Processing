```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_img(title, img):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

img = cv2.imread('Fingerprint.png', 0)
```

```python
show_img('Original Image', img)
```

```python
_, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
kernel = np.ones((3, 3), np.uint8)
eroded = cv2.erode(binary, kernel, iterations=1)
```

```python
show_img('Eroded Image', eroded)
```

```python
dilated = cv2.dilate(binary, kernel, iterations=1)
show_img('Dilated Image', dilated)
```

```python
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
show_img('Opening (Noise Removal)', opening)
```

```python
closing = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)
show_img('Closing (Hole Filling)', closing)
```

```python
operations = {
    'Original': binary,
    'Erosion': cv2.erode(binary, kernel),
    'Dilation': cv2.dilate(binary, kernel),
    'Opening': cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel),
    'Closing': cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel),
}

plt.figure(figsize=(15, 5))
for i, (name, img) in enumerate(operations.items(), 1):
    plt.subplot(1, 5, i)
    plt.imshow(img, cmap='gray')
    plt.title(name)
    plt.axis('off')
plt.show()
```

```python
gradient = cv2.morphologyEx(binary, cv2.MORPH_GRADIENT, kernel)
show_img('Morphological Gradient', gradient)
```

```python
tophat = cv2.morphologyEx(binary, cv2.MORPH_TOPHAT, kernel)
show_img('Top-hat Transformation', tophat)
```

```python
blackhat = cv2.morphologyEx(binary, cv2.MORPH_BLACKHAT, kernel)
show_img('Black-hat Transformation', blackhat)
```

```python
advanced_ops = {
    'Original': binary,
    'Gradient': cv2.morphologyEx(binary, cv2.MORPH_GRADIENT, kernel),
    'Top-hat': cv2.morphologyEx(binary, cv2.MORPH_TOPHAT, kernel),
    'Black-hat': cv2.morphologyEx(binary, cv2.MORPH_BLACKHAT, kernel),
}

plt.figure(figsize=(15, 5))
for i, (name, img) in enumerate(advanced_ops.items(), 1):
    plt.subplot(1, 4, i)
    plt.imshow(img, cmap='gray')
    plt.title(name)
    plt.axis('off')
plt.show()
```
