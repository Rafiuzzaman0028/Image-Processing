# Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Import and read image

```python
image = cv2.imread('E:\Dataset\Alzheimers Dataset ( 4 class of Images)\ModerateDemented\moderateDem2.jpg')
```

### Show Image

```python
plt.imshow(image)
plt.show()
```

### Check Image Shape

```python
print("Shape of the image:", image.shape)
```

### Check Image Matrix

```python
print("Matrix of the image (first 5x5 pixels):")
print(image[:5, :5])
```

### RGB to Gray Scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray_image.shape
```

```python
brightness = np.mean(gray_image)
print(f"Brightness (Average pixel intensity): {brightness}")
```

```python
contrast = np.std(gray_image)
print(f"Contrast (Standard deviation of pixel values): {contrast}")
```

```python
alpha = 2  
beta = 30  
adjusted_image = cv2.convertScaleAbs(gray_image, alpha=alpha, beta=beta)
```

```python
# Plot the original and adjusted images
plt.figure(figsize=(10, 5))

# Plot original image
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('off') 

# Plot adjusted image
plt.subplot(1, 2, 2)
plt.imshow(adjusted_image, cmap='gray')
plt.title('Adjusted Image')
plt.axis('off') 

plt.show()
```

```python
# Calculate the histogram of the image
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# Plot the histogram
plt.plot(histogram)
plt.title('Histogram of Grayscale Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Apply Histogram Equalization

equalized_image = cv2.equalizeHist(gray_image)

# Calculate the histogram of the equalized image
equalized_histogram = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

# Plot the images and histograms in a 2x2 grid
plt.figure(figsize=(12, 12))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Equalized Image
plt.subplot(2, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

# Original Histogram
plt.subplot(2, 2, 3)
plt.plot(histogram)
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# Equalized Histogram
plt.subplot(2, 2, 4)
plt.plot(equalized_histogram)
plt.title('Histogram of Equalized Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# Display the plots
plt.tight_layout()
plt.show()
```

## Apply Gamma correction

```python
gamma = 3.2 
gamma_corrected_image = np.array(255 * (gray_image / 255) ** gamma, dtype='uint8')
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected_image, cmap='gray')
plt.title('Gamma Corrected Image')
plt.axis('off')

plt.show()
```
